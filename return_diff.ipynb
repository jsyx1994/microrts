```python
import pandas as pd
import numpy as np
from pylab import *
import seaborn as sns
from mpl_toolkits.axes_grid1 import host_subplot
import mpl_toolkits.axisartist as AA
```

```python
data = pd.read_csv("return_diff.csv")
r = data.Value.rolling(window=150)
```

```python
x = np.array(data.Step.to_list())
y = r.mean()
```

```python
with sns.axes_style("darkgrid"):
    host = host_subplot(111, axes_class=AA.Axes)
    std = r.std()
    host.plot(x, y)
    host.fill_between(x, y-std, y + std, alpha = 0.3,linewidth=0)
    host.set_xlabel("Episodes")
    host.set_ylabel("Rewards Difference")
    draw()
    show()
```
